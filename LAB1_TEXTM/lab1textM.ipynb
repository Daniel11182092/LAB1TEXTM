## **Problema 1**

```python
imgto = cv2.imread("/content/Paisaje.jpg")
def colores(color,img):
  alto = img.shape[0]
  ancho = img.shape[1]
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  lienzo = np.zeros((alto, ancho, 3))
  for i in range(0, alto):
    for j in range(0, ancho):
        pixel = img[i, j]
        blue = pixel[2]
        green = pixel[1]
        red = pixel[0]
        if color == 1:
          lienzo[i, j] = [blue, 0, 0]
        elif color == 2:
          lienzo[i, j] = [0, green, 0]
        elif color == 3:
          lienzo[i, j] = [0, 0, red]
        elif color == 10:
          lienzo[i, j] = [0, green, red]
        elif color == 20:
          lienzo[i, j] = [blue, green, 0]
        elif color == 30:
          lienzo[i, j] = [blue, 0, red]
  lienzo[i, j] = int(0.299*blue + 0.587*green + 0.114*red)
  cv2.imwrite("monocromo.jpg", lienzo)
  monocromo = cv2.imread("monocromo.jpg")
  monocromo = cv2.cvtColor(monocromo, cv2.COLOR_BGR2RGB)
  plt.imshow(monocromo) 
  return plt.show()
```

```python
colores (1,imgto)
```

```python
imgto = cv2.imread("/content/perro2.jpg")
def colores(img):
  alto = img.shape[0]
  ancho = img.shape[1]
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  lienzo = np.zeros((alto, ancho, 3))
  for i in range(0, alto):
    for j in range(0, ancho):
        pixel = img[i, j]
        blue = pixel[2]
        green = pixel[1]
        red = pixel[0]
        l1=lienzo[i, j] = [blue, 0, 0]
        l2=lienzo[i, j] = [0, green, 0]
        l3=lienzo[i, j] = [0, 0, red]
  cv2.imwrite("img1.jpg", l1)
  img1 = cv2.imread("img1.jpg")
  img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
  cv2.imwrite("img2.jpg", l2)
  img2 = cv2.imread("img2.jpg")
  img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
  cv2.imwrite("img3.jpg", l3)
  img3 = cv2.imread("img3.jpg")
  img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
  plt.imshow(img1, img2 , img3) 
  return plt.show()
```

```python
colores(imgto)
```

```python
imgto = cv2.imread("/content/perro2.jpg")
def histoimg (img) :
  color = ('b','g','r')
  for i, c in enumerate(color):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color = c)
    plt.xlim([0,256])
  
  return plt.show()

  
```

```python
histoimg(imgto)
```
