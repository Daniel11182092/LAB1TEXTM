```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## **Problema 1**

```python
imgto = cv2.imread("/content/Paisaje.jpg")
def colores(color,img):
  alto = img.shape[0]
  ancho = img.shape[1]
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  lienzo = np.zeros((alto, ancho, 3))
  for i in range(0, alto):
    for j in range(0, ancho):
        pixel = img[i, j]
        blue = pixel[2]
        green = pixel[1]
        red = pixel[0]
        if color == 1:
          lienzo[i, j] = [blue, 0, 0]
        elif color == 2:
          lienzo[i, j] = [0, green, 0]
        elif color == 3:
          lienzo[i, j] = [0, 0, red]
        elif color == 10:
          lienzo[i, j] = [0, green, red]
        elif color == 20:
          lienzo[i, j] = [blue, green, 0]
        elif color == 30:
          lienzo[i, j] = [blue, 0, red]
  lienzo[i, j] = int(0.299*blue + 0.587*green + 0.114*red)
  cv2.imwrite("monocromo.jpg", lienzo)
  monocromo = cv2.imread("monocromo.jpg")
  monocromo = cv2.cvtColor(monocromo, cv2.COLOR_BGR2RGB)
  plt.imshow(monocromo) 
  return plt.show()
```

```python
colores (10,imgto)
```
